```
np_x_list.shape[0]
np_x_list[0, :] - np_y
```

```
def euclidean(x, y):
  minimal_value = x[0 , 1:] - y[1:]
  minimal_value = minimal_value * minimal_value * minimal_value * minimal_value
  minimal_value = np.sqrt(minimal_value)
  minimal_value_sum = minimal_value.sum() 
  index = -1
  i = 0

  for temp in x:
    
    temp = temp[1:] - y[1:]
    temp = temp * temp
    temp = np.sqrt(temp)
    temp_value_sum = temp.sum()
    
    if (temp_value_sum < minimal_value_sum):
      minimal_value_sum = temp_value_sum
      index = i
    i += 1
    

  print('가장 비슷한 행은 = ', index)
  print('가장 작은 Euclidean Distance 값은 : ', minimal_value_sum)

  return index
```

```
minimal_index = euclidean(np_x_list, np_y)
```

```
plt.plot(np_x_list[minimal_index, 1:])
plt.plot(np_y[1:])
plt.show()
```

```
def euclidean(x, y):
  maxl_value_sum = 0
  index = -1
  i = 0

  for temp in x:
    
    temp = temp[1:] - y[1:]
    temp = temp * temp
    temp = np.sqrt(temp)
    temp_value_sum = temp.sum()
    
    if (temp_value_sum > maxl_value_sum):
      maxl_value_sum = temp_value_sum
      index = i
    i += 1
    

  print('가장 다른 행은 = ', index)
  print('가장 큰 Euclidean Distance 값은 : ', maxl_value_sum)

  return index
```

```
max_index = euclidean(np_x_list, np_y)
```

```
plt.plot(np_x_list[max_index, 1:])
plt.plot(np_y[1:])
plt.show()
```
