```
data = np.copy(LSF5_10[ :1028])
data = data.reshape(-1,)
data.shape
```

```
def PAA(data, windowSize):
  arr = np.array([])

  for i in range(0, data.size, windowSize):
    temp = np.mean(data[i:i+windowSize])
    arr = np.append(arr, [temp])

  return arr
```

```
temp = PAA(data, 10)
```

```
temp
```

```
plt.subplot(2,1,1)
plt.title("original data")
plt.plot(data)
plt.grid()

plt.subplot(2,1,2)
plt.title("functioned PAA")
plt.plot(temp)
plt.grid()

plt.show()
```
